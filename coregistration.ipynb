# Data coregistration using Arosics

### Global coregistration

```python
from arosics import COREG

im_reference = ''
im_target = ''

path_out = ''

CR = COREG(im_reference, im_target, path_out=path_out, fmt_out='ISIS3', ws=(1000, 1000), max_shift=900, nodata=(None, None), match_gsd=True, s_b4match=3)
CR.calculate_spatial_shifts()

# CR.show_matchWin()

CR.correct_shifts()
```

### Local coregistration

```python
from arosics import COREG_LOCAL

im_reference = ''
im_target = ''

path_out = ''

kwargs = {
    'grid_res' : 300,
    'path_out' : path_out,
    'fmt_out' : 'ENVI',
    'window_size' : (800, 800),
    'max_shift' : 50,
    'nodata' : (None, 65535),
    'match_gsd' : True,
}

CRL = COREG_LOCAL(im_reference, im_target, **kwargs)
CRL.correct_shifts()
```

### Batch global coregistration

```python
import os
from arosics import COREG
from utils import holden_crater_pairs, get_base_name


evc_folder = ""
ctx_folder = ""
output_folder = ""

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(evc_folder):
    try:
        if file.endswith('.img'):

            evc_base_name = file.split('.')[0]

            ctx_base_name = holden_crater_pairs[evc_base_name]

            evc_path = os.path.join(evc_folder, file)
            ctx_path = os.path.join(ctx_folder, f"{ctx_base_name}.cub")

            output_path = os.path.join(output_folder, f"{evc_base_name}.rp.global.img")

            CR = COREG(ctx_path, evc_path, path_out=output_path, fmt_out='ENVI', ws=(1200, 1200), max_shift=100, nodata=(None, 65535), match_gsd=True)
            CR.calculate_spatial_shifts()

            CR.correct_shifts()
            print('\n')

    except Exception as e:
        print(f"Error processing {file}: {e}\n")
```

### Batch local coregistration 

```python
import os
from arosics import COREG_LOCAL
from utils import holden_crater_pairs, copy_crs_from_ctx, get_base_name

evc_folder = ""
ctx_folder = ""
output_folder = ""

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(evc_folder):
    try:
        if file.endswith('.img'):

            evc_base_name = get_base_name(file)

            ctx_base_name = holden_crater_pairs[evc_base_name]

            evc_path = os.path.join(evc_folder, file)
            ctx_path = os.path.join(ctx_folder, f"{ctx_base_name}.cub")

            output_path = os.path.join(output_folder, f"{evc_base_name}.rp.global.local.img")

            kwargs = {
                'grid_res' : 300,
                'path_out' : output_path,
                'fmt_out' : 'ENVI',
                'window_size' : (800, 800),
                'max_shift' : 50,
                'match_gsd' : True,
            }

            CRL = COREG_LOCAL(im_ref=ctx_path, im_tgt=evc_path, **kwargs)
            CRL.correct_shifts()
            print('\n')


    except Exception as e:
        print(f"Error processing {file}: {e}")
        print('\n')
```
